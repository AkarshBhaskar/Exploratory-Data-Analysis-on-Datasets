### Dropping Unique Columns

```python
df.drop(['Unnamed: 0','id'],axis=1,inplace=True)
```

### Dropping columns having a single class

```python
df.drop(['requires_license','is_business_travel_ready'],axis=1,inplace=True)
```

```python
df['cleaning_fee']=df['cleaning_fee'].astype('str')
x=df['cleaning_fee'].to_list()
for i in range(len(x)):
    x[i]=x[i].strip('$')
    x[i]=x[i].replace(',','')
    if x[i]=='nan'or x[i]=='NaN':
        x[i]=np.nan
df['cleaning_fee']=pd.DataFrame(x)        
df['cleaning_fee']=pd.to_numeric(df['cleaning_fee'])
```

```python
df['security_deposit']=df['security_deposit'].astype('str')
x=df['security_deposit'].to_list()
for i in range(len(x)):
    x[i]=x[i].strip('$')
    x[i]=x[i].replace(',','')
    if x[i]=='nan'or x[i]=='NaN':
        x[i]=np.nan
df['security_deposit']=pd.DataFrame(x)        
df['security_deposit']=pd.to_numeric(df['security_deposit'])
```

```python
df.info()
```

```python
num_cols=df.select_dtypes(include=['int64','float64']).columns
cat_cols=df.select_dtypes(include='object').columns
num_cols,cat_cols
```

### Finding Null columns

```python
num_null_cols=[]
cat_null_cols=[]
for i in num_cols:
    if df[i].isnull().sum()>0:
        num_null_cols.append(i)
for i in cat_cols:
    if df[i].isnull().sum()>0:
        cat_null_cols.append(i)
num_null_cols,cat_null_cols
```

### Finding outlier columns

```python
outlier_cols=[]
for i in num_cols:
    iqr=df[i].quantile(0.75)-df[i].quantile(0.25)
    higher_range=df[i].quantile(0.75)+(1.5*iqr)
    lower_range=df[i].quantile(0.25)-(1.5*iqr)
    if df[i][df[i]<lower_range].count()>0 or df[i][df[i]>higher_range].count()>0:
        outlier_cols.append(i)
outlier_cols
```

### Numerical Null Imputation

```python
for i in num_null_cols:
    if i in outlier_cols:
        df[i]=df[i].fillna(df[i].median())
    else:
        df[i]=df[i].fillna(df[i].mean())
```

```python
for i in num_null_cols:
    if df[i].isnull().sum()>0:
        print(True)
    else:
        print(False)
```

### Categorical Null Imputation

```python
for i in cat_null_cols:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
for i in cat_null_cols:
    if df[i].isnull().sum()>0:
        print(True)
    else:
        print(False)
```

```python
df[outlier_cols]
```

### Outlier Imputation

```python
for i in outlier_cols:
    iqr=df[i].quantile(0.75)-df[i].quantile(0.25)
    higher_range=df[i].quantile(0.75)+(1.5*iqr)
    lower_range=df[i].quantile(0.25)-(1.5*iqr)
    replace_higher_value=df[i].quantile(0.9)
    replace_lower_value=df[i].quantile(0.1)
    outlier_treated=np.where(df[i]>higher_range,replace_higher_value,df[i])
    df[i]=pd.DataFrame(outlier_treated)
    outlier_treated=np.where(df[i]<lower_range,replace_lower_value,df[i])
    df[i]=pd.DataFrame(outlier_treated)
```

```python
df[outlier_cols]
```

```python
new_outlier_cols=[]
for i in outlier_cols:
    iqr=df[i].quantile(0.75)-df[i].quantile(0.25)
    higher_range=df[i].quantile(0.75)+(1.5*iqr)
    lower_range=df[i].quantile(0.25)-(1.5*iqr)
    if df[i][df[i]<lower_range].count()>0 or df[i][df[i]>higher_range].count()>0:
        new_outlier_cols.append(i)
        print(True,i)
    else:
        print(False)
new_outlier_cols
```

```python
for i in new_outlier_cols:
    iqr=df[i].quantile(0.75)-df[i].quantile(0.25)
    higher_range=df[i].quantile(0.75)+(1.5*iqr)
    lower_range=df[i].quantile(0.25)-(1.5*iqr)
    replace_higher_value=df[i].quantile(0.8)
    replace_lower_value=df[i].quantile(0.2)
    outlier_treated=np.where(df[i]>higher_range,replace_higher_value,df[i])
    df[i]=pd.DataFrame(outlier_treated)
    outlier_treated=np.where(df[i]<lower_range,replace_lower_value,df[i])
    df[i]=pd.DataFrame(outlier_treated)
```

### Encoding

```python
for i in cat_cols:
    if df[i].value_counts().count()<3:
        df[i]=df[i].map({'t':1,'f':0})   
    else:
        df[i]=LabelEncoder().fit_transform(df[i])
```
