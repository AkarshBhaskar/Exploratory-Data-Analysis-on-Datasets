```python
df.drop(['Unnamed: 0','id'],axis=1,inplace=True)
```

```python
df['cleaning_fee']=df['cleaning_fee'].astype('str')
x=df['cleaning_fee'].to_list()
for i in range(len(x)):
    x[i]=x[i].strip('$')
    x[i]=x[i].replace(',','')
df['cleaning_fee']=pd.DataFrame(x)        
df['cleaning_fee'].astype('float64')
```

```python
df['security_deposit']=df['security_deposit'].astype('str')
for i in range(len(df['security_deposit'])):
    df['security_deposit'][i]=df['security_deposit'][i].strip('$')
    df['security_deposit'][i]=df['security_deposit'][i].replace(',','')
    if df['security_deposit'][i]=='Nan' or df['security_deposit'][i]=='nan':
        df['security_deposit'][i]=np.nan
df['security_deposit'].astype('float64')
```
