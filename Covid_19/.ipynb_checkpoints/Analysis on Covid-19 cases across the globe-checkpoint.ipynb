```python
#13 write a function to calculate standard deviation. (don't use builtin methods)
```

```python
def stdev(col):
    
    lst=col.tolist()
    mean=sum(lst)/len(lst)
    variance=0
    for i in lst:
        variance+=((i-mean)**2)
    variance=variance/len(lst)
    return variance**0.5

col=df['Confirmed']
stdev(col)    
```

```python
#14 write a function to calculate pearson's r for two numerical columns.
```

```python
df[['Confirmed','Recovered']].corr()
```

```python
def pcorr(col1,col2):
    
    col1=col1.tolist()
    col2=col2.tolist()
    mean1=sum(col1)/len(col1)
    mean2=sum(col2)/len(col2)
    x,y,z=0,0,0
    for i in range(len(col1)):
        x+=(col1[i]-mean1)*(col2[i]-mean2)
        y+=(col1[i]-mean1)**2
        z+=(col2[i]-mean2)**2
    return x/((y*z)**0.5)

pcorr(df['Confirmed'],df['Recovered'])
```
